```python
import pandas as pd #data manipulation
import numpy as np #maths + stats
import matplotlib.pyplot as plt #visualization
import seaborn as sns #visualization
```

```python
orders = pd.read_csv('Orders (1).csv')
customers = pd.read_csv('Customers.csv')
restaurants = pd.read_csv('Restaurants.csv')
```

```python
orders.head()
```

```python
orders.info()
```

```python
orders['Order_Date'] = pd.to_datetime(orders['Order_Date'])
```

```python
orders.info()
```

```python
orders
```

```python
orders['Month'] = orders['Order_Date'].dt.strftime("%B")
```

```python
orders
```

```python
plt.figure(figsize=(6,3))
sns.countplot(x = "Month", data = orders, hue = "Month", palette='spring') #palette = viridis
plt.xticks(rotation = 90)
plt.show()
```

```python
orders['Hour'] = orders['Order_Date'].dt.strftime("%H")
orders['Hour'] = orders['Hour'].astype('int')
orders = orders.sort_values(by = 'Hour')
sns.lineplot(x = 'Hour', y = 'Total_Amount', data = orders, estimator = "sum")
plt.show()
```

```python
customers
```

```python
df = pd.merge(left = orders, right = customers, on = 'Customer_ID', how = "inner")
```

```python
df
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x = 'Customer_Location', data = df, hue = 'Order_Status')
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x = 'Customer_Location', data = df, hue = 'Dish_Name')
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x = 'Customer_Location', data = df, hue = 'Dish_Name', palette='viridis', edgecolor = 'black')
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x = 'Customer_Location', data = df, hue = 'Customer_Age_Group', palette='viridis', edgecolor = 'black')
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```
